Com a analise exploratoria feita no arquivo Analise Exploratoria, vamos para parte de visualização de Dados

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Banco_de_Dados = pd.read_csv('bank-additional/bank-additional/bank-additional-full.csv', delimiter=';')
```

```python
Banco_de_Dados
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(Banco_de_Dados.drop(['y'],axis=1).select_dtypes(include=['int','float']).columns):
    ax = plt.subplot(4,3, i+1)
    sns.histplot(data=Banco_de_Dados, x=col, ax=ax,color='red',kde=True, hue ='y')
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
```

```python
plt.figure(figsize=(24, 24))
for i, col in enumerate(Banco_de_Dados.drop(['y'],axis=1).select_dtypes(exclude=['int','float']).columns):
    ax = plt.subplot(5,2, i+1)
    sns.histplot(data=Banco_de_Dados, x=col, ax=ax,color='red', hue ='y')
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
```

```python
plt.figure(figsize=(18, 18))
for i, col in enumerate(Banco_de_Dados.drop(['y'],axis=1).select_dtypes(include=['int','float']).columns):
    ax = plt.subplot(4,3, i+1)
    sns.boxplot(data=Banco_de_Dados, x=col, ax=ax,color='red')
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
```
